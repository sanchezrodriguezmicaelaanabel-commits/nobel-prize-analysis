## 📊 Resumen del Análisis Exploratorio de Datos (EDA)

Este proyecto permitió realizar un análisis demográfico, de tendencias y de género sobre los ganadores de los Premios Nobel y su historia desde su creación en 1901 hasta 2023, utilizando herramientas clave como **Pandas, Seaborn y NumPy**.

A partir del análisis, se desprenden los siguientes hallazgos principales:

* **Hegemonía geográfica e histórica:** El análisis exploratorio demuestra que los galardonados nacidos en los Estados Unidos y el país en sí mismo dominan históricamente la obtención de premios. Su crecimiento comenzó en la posguerra y alcanzó su proporción más alta de ganadores estadounidenses en la década de los 2000.

* **Evolución de la brecha de género:** Los datos revelan un crecimiento lento pero constante en la presencia e inclusión de las mujeres en la academia, registrándose en la década de 2020 la mayor proporción histórica de mujeres laureadas. Según lo observado, la categoría de Literatura es la que mostró la mayor presencia de ganadoras femeninas a lo largo de los años.

* **Figuras femeninas destacadas:** Al filtrar los datos, se destacaron varias figuras femeninas notables, siendo Marie Curie la primera mujer en recibir un Premio Nobel, hito que logró en la categoría de Física.

* **Casos excepcionales:** A través del filtrado de datos, pude notar que varios galardonados recibieron dos o más premios a lo largo de su carrera debido a su gran dedicación e innovación en sus campos de estudio correspondientes.
